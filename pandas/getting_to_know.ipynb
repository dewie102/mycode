```python
import pandas as pd
```

```python
excel_file = 'movies.xls'

excel_file
```

## When data is inserted into Pandas, it is turned into a DATAFRAME

```python
movies_sheet_1 = pd.read_excel(excel_file, sheet_name=0, index_col=0)
```

```python
movies_sheet_1
```

```python
movies_sheet_2 = pd.read_excel(excel_file, sheet_name=1, index_col=0)
```

```python
movies_sheet_2
```

```python
movies_sheet_3 = pd.read_excel(excel_file, sheet_name=2, index_col=0)
```

```python
movies_sheet_3
```

```python
# Combine all DFs into a single DF called movies

movies = pd.concat([movies_sheet_1, movies_sheet_2, movies_sheet_3])
```

```python
movies
```

```python
movies.shape
```

```python
# THERE ARE REPEATS!

movies.drop_duplicates(inplace=True)
```

```python
movies.shape
```

```python
sorted_by_gross = movies.sort_values(["Gross Earnings"], ascending=False)
```

```python
sorted_by_gross.head(10)
```
